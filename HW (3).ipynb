```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/flying-etiquette-survey/flying-etiquette.csv')


age_counts = df['Age'].value_counts()
age_percentages = age_counts / age_counts.sum() * 100

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=age_percentages.index, y=age_percentages.values)
plt.title('Percent Breakdown of Survey Participants by Age Range')
plt.xlabel('Age Range')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Is it rude to ask someone to switch seats with you in order to be closer to friends?')
plt.title('Respondents\' Feelings on Asking to Switch Seats')
plt.xlabel('Feelings')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Is it rude to ask someone to switch seats with you in order to be closer to friends?', hue='Age')
plt.title('Respondents\' Feelings on Asking to Switch Seats by Age')
plt.xlabel('Feelings')
plt.ylabel('Count')
plt.show()
```
